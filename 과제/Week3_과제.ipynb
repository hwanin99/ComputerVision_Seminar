```python
import cv2
from google.colab.patches import cv2_imshow
from PIL import Image as PILImage
import matplotlib.pyplot as plt
import glob

dolphin_img_list = glob.glob('/content/drive/MyDrive/CV_seminar_project/original/dolphin/*')
shark_img_list = glob.glob('/content/drive/MyDrive/CV_seminar_project/original/shark/*')
whale_img_list = glob.glob('/content/drive/MyDrive/CV_seminar_project/original/whale/*')
```

# 과제 1번

```python
for i in range(len(list(range(int(input()))))):
  dolphin_img=cv2.imread(dolphin_img_list[i]) 
  shark_img=cv2.imread(shark_img_list[i])
  whale_img=cv2.imread(whale_img_list[i])
  
  dolphin=cv2.resize(dolphin_img,(255,255)) 
  shark=cv2.resize(shark_img,(255,255))
  whale=cv2.resize(whale_img,(255,255))
  
  plt.figure(figsize=(10,10))

  plt.subplot(331,title='dolphin',xlabel=dolphin_img.shape)
  plt.imshow(dolphin[:,:,::-1])

  plt.subplot(332,title='shark',xlabel=shark_img.shape)
  plt.imshow(shark[:,:,::-1])
  
  plt.subplot(333,title='whale',xlabel=whale_img.shape)
  plt.imshow(whale[:,:,::-1])
```

# 과제 2번

```python
img_2=cv2.imread(dolphin_img_list[0])
img_2.shape # (183,275,3)

cv2_imshow(img_2) 
```

```python
(h,w,c)=img_2.shape

for i in range(20,w-20):
  img_2[h//6:h//6+5,i]=0 # Black
  img_2[(h//6+h//2)//2:(h//6+h//2)//2+5,i]=[0,0,255] # Red
  img_2[h//2:h//2+5,i]=[0,255,0] # Green
  img_2[(h-(h//6)+h//2)//2:(h-(h//6)+h//2)//2+5,i]=[255,0,0] # Blue
  img_2[h-(h//6):h-(h//6)+5,i]=255 # White
cv2_imshow(img_2)
```
